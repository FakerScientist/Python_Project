```python
#define function could be nice for future data
def parse_string_list(pd_listlike_object):
    """
    
    args: a pandas series object that looks like a list of strings
    returns: a list of strings
    
    """
    return [s for s in pd_listlike_object.strip("[]").split(',')]
```

```python
def parsefunc(stringdata):
    strippeddata = stringdata.strip("[]")
    splitdata = strippeddata.split(',')
    return [x.strip().strip("'") for x in splitdata]
    
```
