# Data cleaning for cleandf

```python
#Cleaning heroes column into list of strings
def parsefunc(stringdata:str) -> List[str]:
    """
    Function for changing pandas object into a list.
    
    arg: column of a data frame that is pd object
    
    return: list of strings from that column 
    """
    strippeddata = stringdata.strip("[]")
    splitdata = strippeddata.split(',')
    return [x.strip().strip("'") for x in splitdata]  
```

```python
cleandf['TEAM_1_HEROES'] = cleandf['TEAM_1_HEROES'].apply(lambda x: parsefunc(x))
cleandf['TEAM_1_HEROES']
```

```python
cleandf['TEAM_0_HEROES'] = cleandf['TEAM_0_HEROES'].apply(lambda x: parsefunc(x))
cleandf['TEAM_0_HEROES']
```

```python
#Function for adding index column where index is the same number for each game.
index = 0
list_ = []
for item in cleandf.TEAM_0_HEROES:
    list_.extend(map(lambda hero: [index, hero], item))
    index += 1
hero = pd.DataFrame(list_, columns=['index', 'TEAM_0_HEROES'])
hero.head(15)
```

```python
#Function for adding index column where index is the same number for each game.
index = 0
list_ = []
for item in cleandf.TEAM_1_HEROES:
    list_.extend(map(lambda hero: [index, hero], item))
    index += 1
hero1 = pd.DataFrame(list_, columns=['index1', 'TEAM_1_HEROES'])
hero1.head(15)
```

```python
#The dataframes can then be combined on the index column.
hero2 = pd.concat([hero, hero1], axis = 1)
hero2 = hero2.drop('index1', axis=1)
hero2
```

```python
#Then the old heroes columns are dropped to allow for the new ones to be added in with the merge.
cleandf_allheroes_merged = pd.merge(cleandf.drop(['TEAM_0_HEROES', 'TEAM_1_HEROES'], axis=1), 
                                    hero2, 
                                    how='right', 
                                    left_index=True, 
                                    right_on='index'
                                   )
cleandf_allheroes_merged.head(20)
```

# Creating Functions, DataFrames, and Feature Engineering

```python
#function for determining winrate and enabling feature engineering for winrates
hero_win_rate = {}

def win_rate(row):
    """
    Docstring:
    This function needs a dictionary to be initialized before use. 
    It is used to calculate the winrate for every hero in the dataframe.
    
    :args: It says row, but its really a dataframe.
    
    :return: A dictionary. The key of the dictionary is the hero name and the
    value is a tuple. The first integer in the tuple is a count of how many games the hero won, and the
    second integer is how many games a hero was played in total.
    """
    
    if row['TEAM_0_WIN'].any():
        heroes = row['TEAM_0_HEROES']
        for hero in heroes:
            if hero in hero_win_rate:
                hero_win_rate[hero][0] += 1
            elif hero in hero_win_rate:
                hero_win_rate[hero] = [1, 0]
            else:
                heroes = row['TEAM_0_HEROES']
        for hero in heroes:
            if hero in hero_win_rate:
                hero_win_rate[hero][1] += 1
            else:
                hero_win_rate[hero] = [0, 1]

    if row['TEAM_1_WIN'].any():
        heroes = row['TEAM_1_HEROES']
        for hero in heroes:
            if hero in hero_win_rate:
                hero_win_rate[hero][0] += 1
            elif hero in hero_win_rate:
                hero_win_rate[hero] = [1, 0]
            else:
                heroes = row['TEAM_1_HEROES']
        for hero in heroes:
            if hero in hero_win_rate:
                hero_win_rate[hero][1] += 1
            else:
                hero_win_rate[hero] = [0, 1]
```

```python
#Tool for seeing how often heroes are picked with each other.
pairs = Counter()
for row in cleandf['TEAM_1_HEROES']:
    for pair in combinations(row, 2): #change this number for how many heroes you want to see grouped
        pairs[pair] += 1
for row in cleandf['TEAM_0_HEROES']:
    for pair in combinations(row, 2): #change this number for how many heroes you want to see grouped
        pairs[pair] += 1

top_pairs = pd.DataFrame(pairs.most_common(), columns = ['Heroes', 'Games Together'])
top_pairs = top_pairs.set_index('Heroes')
top_pairs
```
