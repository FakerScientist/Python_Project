```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('tier_1_2019_2022.csv')

# Create a pivot table with the count of wins and losses for each tournament
table = pd.pivot_table(df, index=['TOURNAMENT'], columns=['RESULT'], aggfunc='count')
table.reset_index(inplace=True)
table.rename(columns={'WIN': 'WIN', 'LOSE': 'LOSE'}, inplace=True)

# Calculate the ratio of wins to losses
table['Win Ratio'] = table['WIN'] / (table['WIN'] + table['LOSE'])

# Create the grouped bar chart
sns.set(style='whitegrid')
sns.set_palette("pastel")
plt.figure(figsize=(10, 6))
sns.barplot(data=table, x='TOURNAMENT', y='Win Ratio', color='b')
plt.title('Win Ratio by Tournament')
plt.xlabel('Tournament')
plt.ylabel('Win Ratio')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('tier_1_2019_2022.csv')

# Group the data by tournament and result and get the count of each group
table = df.groupby(['TOURNAMENT', 'RESULT']).size().unstack(level=-1, fill_value=0).reset_index()

# Calculate the ratio of wins to losses
table['Win Ratio'] = table['WIN'] / (table['WIN'] + table['LOSE'])

# Create the grouped bar chart
sns.set(style='whitegrid')
sns.set_palette("pastel")
plt.figure(figsize=(10, 6))
sns.barplot(data=table, x='TOURNAMENT', y='Win Ratio', color='b')
plt.title('Win Ratio by Tournament')
plt.xlabel('Tournament')
plt.ylabel('Win Ratio')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('tier_1_2019_2022.csv')

# Group the data by tournament and result and get the count of each group
table = df.pivot_table(index='TOURNAMENT', columns='RESULT', values='MATCH_ID', aggfunc='count', fill_value=0).reset_index()

# Calculate the ratio of wins to losses
table['Win Ratio'] = table['WIN'] / (table['WIN'] + table['LOSE'])

# Create the grouped bar chart
sns.set(style='whitegrid')
sns.set_palette("pastel")
plt.figure(figsize=(10, 6))
sns.barplot(data=table, x='TOURNAMENT', y='Win Ratio', color='b')
plt.title('Win Ratio by Tournament')
plt.xlabel('Tournament')
plt.ylabel('Win Ratio')
plt.xticks(rotation=45)
plt.show()
```
